```python
# Prepare survey data across multiple waves for modeling
# This loop creates datasets for each wave (4,5,6,8), removes incomplete stress data,
# encodes categorical variables, and combines them into a single dataframe
from sklearn.preprocessing import OrdinalEncoder

waves = [4,5,6,8]

# Map survey wave numbers to PSQI column indices
# Waves are non-sequential (4,5,6,8) but PSQI columns are sequential (1,2,3,4)
wave_map = {4:1, 5:2, 6:3, 8:4}

datasets = []

for wave in waves:

    suffix = wave_map[wave]

    df = pd.DataFrame({
        "stress": survey[f"Stress_{wave}"],
        "happy": survey[f"Happy_{wave}"],
        "health": survey[f"Health_{wave}"],
        "trust": survey[f"Trust_{wave}"],
        "sleep_quality": survey[f"PSQIGlobal_{suffix}"],
        "sedentary": survey[f"sedentarytime_scale_{wave}"],
        "exercise_alone": survey[f"exercisealone_{wave}"],
        "exercise_others": survey[f"exerciseothers_{wave}"],
        "beer": survey[f"usebeer_{wave}"],
        "wine": survey[f"usewine_{wave}"],
        "caffeine": survey[f"usecaffine_{wave}"],
        "tobacco": survey[f"usetobacco_{wave}"]
    })

    df["wave"] = wave

    # remove rows without stress
    df = df.dropna(subset=["stress"])

    categorical_cols = [
        "happy","health","exercise_alone","exercise_others",
        "beer","wine","caffeine","tobacco"
    ]

    encoder = OrdinalEncoder()
    df[categorical_cols] = encoder.fit_transform(df[categorical_cols])

    datasets.append(df)

survey_model_data = pd.concat(datasets, ignore_index=True)

survey_model_data.shape
```
