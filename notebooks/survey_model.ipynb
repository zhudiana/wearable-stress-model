```python
# Prepare survey data across multiple waves for modeling
# This loop creates datasets for each wave (4,5,6,8), removes incomplete stress data,
# encodes categorical variables, and combines them into a single dataframe

from sklearn.preprocessing import OrdinalEncoder

waves = [4,5,6,8]

# Map survey wave numbers to PSQI column indices
wave_map = {4:1, 5:2, 6:3, 8:4}

datasets = []

for wave in waves:

    suffix = wave_map[wave]

    df = pd.DataFrame({
        "egoid": survey["egoid"],   # <-- ADD THIS (student ID)

        "stress": survey[f"Stress_{wave}"],
        "happy": survey[f"Happy_{wave}"],
        "health": survey[f"Health_{wave}"],
        "trust": survey[f"Trust_{wave}"],
        "sleep_quality": survey[f"PSQIGlobal_{suffix}"],
        "sedentary": survey[f"sedentarytime_scale_{wave}"],
        "exercise_alone": survey[f"exercisealone_{wave}"],
        "exercise_others": survey[f"exerciseothers_{wave}"],
        "beer": survey[f"usebeer_{wave}"],
        "wine": survey[f"usewine_{wave}"],
        "caffeine": survey[f"usecaffine_{wave}"],
        "tobacco": survey[f"usetobacco_{wave}"]
    })

    df["wave"] = wave

    # remove rows without stress
    df = df.dropna(subset=["stress"])

    categorical_cols = [
        "happy","health","exercise_alone","exercise_others",
        "beer","wine","caffeine","tobacco"
    ]

    encoder = OrdinalEncoder()
    df[categorical_cols] = encoder.fit_transform(df[categorical_cols])

    datasets.append(df)

survey_model_data = pd.concat(datasets, ignore_index=True)

survey_model_data.shape
```

```python
from sklearn.model_selection import GroupShuffleSplit
```

```python
X = survey_model_data.drop(columns=["stress"])
y = survey_model_data["stress"]
groups = survey_model_data["egoid"]
```

```python
gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

train_idx, test_idx = next(gss.split(X, y, groups))

X_train = X.iloc[train_idx]
X_test = X.iloc[test_idx]

y_train = y.iloc[train_idx]
y_test = y.iloc[test_idx]
```

```python
train_students = set(X_train["egoid"])
test_students = set(X_test["egoid"])

train_students.intersection(test_students)
```

```python
X_train = X_train.drop(columns=["egoid"])
X_test = X_test.drop(columns=["egoid"])
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.dummy import DummyRegressor
from sklearn.metrics import mean_absolute_error, r2_score

dummy = DummyRegressor(strategy="mean")
dummy.fit(X_train, y_train)

dummy_pred = dummy.predict(X_test)

dummy_mae = mean_absolute_error(y_test, dummy_pred)
dummy_r2 = r2_score(y_test, dummy_pred)

print("Dummy MAE:", dummy_mae)
print("Dummy R2:", dummy_r2)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=200,
    max_depth=10,
    random_state=42
)

rf.fit(X_train, y_train)

rf_pred = rf.predict(X_test)

rf_mae = mean_absolute_error(y_test, rf_pred)
rf_r2 = r2_score(y_test, rf_pred)

print("Random Forest MAE:", rf_mae)
print("Random Forest R2:", rf_r2)
```

```python
import pandas as pd

feature_importance = pd.Series(
    rf.feature_importances_,
    index=X_train.columns
).sort_values(ascending=False)

feature_importance
```

```python
import matplotlib.pyplot as plt

feature_importance.head(10).plot(kind="bar")
plt.title("Top Survey Predictors of Stress")
plt.ylabel("Importance")
plt.show()
```

```python
from sklearn.inspection import permutation_importance

perm = permutation_importance(
    rf,
    X_test,
    y_test,
    n_repeats=10,
    random_state=42
)

perm_importance = pd.Series(
    perm.importances_mean,
    index=X_test.columns
).sort_values(ascending=False)

perm_importance
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, r2_score

xgb = XGBRegressor(
    n_estimators=300,
    max_depth=4,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

xgb.fit(X_train, y_train)

pred = xgb.predict(X_test)

mae = mean_absolute_error(y_test, pred)
r2 = r2_score(y_test, pred)

print("XGBoost MAE:", mae)
print("XGBoost R2:", r2)
```

```python
import pandas as pd

importance = pd.Series(
    xgb.feature_importances_,
    index=X_train.columns
).sort_values(ascending=False)

importance
```

```python
import matplotlib.pyplot as plt

pred = xgb.predict(X_test)

plt.scatter(y_test, pred, alpha=0.5)
plt.xlabel("Actual Stress")
plt.ylabel("Predicted Stress")
plt.title("Actual vs Predicted Stress (XGBoost)")
plt.show()
```

```python
import seaborn as sns

errors = y_test - pred

sns.histplot(errors, bins=30)
plt.title("Prediction Error Distribution")
plt.xlabel("Error (Actual - Predicted)")
plt.show()
```
